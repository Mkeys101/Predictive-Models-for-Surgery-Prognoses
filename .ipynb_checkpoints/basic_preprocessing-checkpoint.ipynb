# Preprocessing
This script contains loading, basic pre-processing, and organisation of the data.

The tasks completed are as follows: 
- Load patient and provider data
- Organise and fix columns of the provider data 
- Organise and fix columns of the patient data
- Left merge the yearly patient level data with provider data 
- Create year columns
- Merge the yearly data 
- Concatenate the data into datasets by procedure type 
- Subset datasets on the features we will use in the most elaborate model
- Save datasets

```python
# Organise column names of patient data 
knee_data = [data_2010_knee, data_2011_knee, data_2012_knee, 
             data_2013_knee, data_2014_knee, data_2015_knee, data_2016_knee]

hip_data = [data_2010_hip, data_2011_hip, data_2012_hip,
            data_2013_hip, data_2014_hip, data_2015_hip, data_2016_hip]

groin_data = [data_2010_groin, data_2011_groin, data_2012_groin,
              data_2013_groin, data_2014_groin, data_2015_groin, data_2016_groin]

vein_data = [data_2010_vein, data_2011_vein, data_2012_vein, 
             data_2013_vein, data_2014_vein, data_2015_vein, data_2016_vein]

knee_columns = {}
hip_columns = {}
groin_columns = {} 
vein_columns = {}

year = 2010

procedures = [knee_data, hip_data, groin_data, vein_data]
```

```python
# Store column names in dictionaries 
for k in range(4):
    i = 0 
    for j in range(len(procedures[k])):
        if k == 0:
            knee_columns[year+i] = list(procedures[k][j])
        elif k == 1: 
            hip_columns[year+i] = list(procedures[k][j])
        elif k == 2: 
            groin_columns[year+i] = list(procedures[k][j])
        else:
            vein_columns[year+i] = list(procedures[k][j])
        i += 1
```

```python
# Python dict does not have a name attribute
knee_columns["name"] = "knee"
hip_columns["name"] = "hip"
groin_columns["name"] = "groin"
vein_columns["name"] = "vein"
```

```python
for j in [knee_columns, hip_columns, groin_columns, vein_columns]: 
    for i in j.values():
        print(j["name"]+" "+str(len(i)))
```
